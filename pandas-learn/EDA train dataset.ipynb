```python
# Анализ категориальных переменных
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    sns.countplot(y=col, data=df)
    plt.title(f'Частота для {col}')
    plt.show()
```

```python
# Визуализация распределения некоторых числовых признаков
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()

for col in numerical_cols:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[col], bins=50, kde=True)
    plt.title(f'Распределение {col}')
    plt.show()
```

# Работа с датами (из формата `%d/%m/%Y` в формат `YYYY-MM-DD`)

```python
# Преобразование колонки 'Order Date' и 'Ship Date' в формат даты (YYYY-MM-DD)
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d/%m/%Y')
```

```python
# Создание новой колонки 'Month' для извлечения месяца из даты
df['Month'] = df['Order Date'].dt.strftime('%Y-%m')
df['Year'] = df['Order Date'].dt.year
```

```python
df.head()
```

```python
# Создание сводной таблицы с 'Ship Mode' в строках, месяцами в столбцах и суммой продаж 'Sales' как метрикой
pivot_table = df.pivot_table(index='Ship Mode', columns='Month', values='Sales', aggfunc='sum', fill_value=0)

# Вывод сводной таблицы
pivot_table
```

```python
# Создание сводной таблицы с 'Ship Mode' в строках, месяцами в столбцах и суммой продаж 'Sales' как метрикой
pivot_table = df.pivot_table(index='Ship Mode', columns='Year', values='Sales', aggfunc='sum', fill_value=0)

# Округление значений до целых
pivot_table = pivot_table.round(0)

# Вывод сводной таблицы
pivot_table
```
